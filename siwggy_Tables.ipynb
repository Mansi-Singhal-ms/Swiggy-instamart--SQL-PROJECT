```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Create database connection using SQLAlchemy
# engine = create_engine('mysql+pymysql://your_username:your_password@localhost/your_database')

engine = create_engine('mysql+pymysql://root:password@localhost/swiggy_db')
```

### All tables from Swiggy Database

```python
customers="""
SELECT * from customers
"""

df_customers = pd.read_sql(customers,engine)
df_customers.head()
```
